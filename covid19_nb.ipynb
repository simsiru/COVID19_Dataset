1) Case Data
- **Case**: Data of COVID-19 infection cases in South Korea

### Case data cleaning

Data of COVID-19 infection cases in South Korea
- case_id: the ID of the infection case
case_id(7) = region_code(5) + case_number(2)
You can check the region_code in 'Region.csv'
- province: Special City / Metropolitan City / Province(-do)
- city: City(-si) / Country (-gun) / District (-gu)
The value 'from other city' means that where the group infection started is other city.
- group: TRUE: group infection / FALSE: not group
If the value is 'TRUE' in this column, the value of 'infection_cases' means the name of group.
The values named 'contact with patient', 'overseas inflow' and 'etc' are not group infection.
- infection_case: the infection case (the name of group or other cases)
The value 'overseas inflow' means that the infection is from other country.
The value 'etc' includes individual cases, cases where relevance classification is ongoing after investigation, and cases under investigation.
- confirmed: the accumulated number of the confirmed
- latitude: the latitude of the group (WGS84)
- longitude: the longitude of the group (WGS84)

```python
case_df = pd.read_csv('Case.csv',
                      index_col=0)
case_df.head()
```

```python
def print_summary_statistics(df: pd.DataFrame) -> None:
  if not df.select_dtypes('number').empty:
    print(df.select_dtypes('number').describe(),'\n')

  if not df.select_dtypes('datetime').empty:
    print(df.select_dtypes('datetime').describe(),'\n')

  if not df.select_dtypes('object').empty:
    print(df.select_dtypes('object').describe())
```

```python
print_summary_statistics(case_df)
```

```python
case_df.shape
```

```python
def print_unique_val_columns(df: pd.DataFrame) -> None:
  for col_name in df.columns:
    print(f'Column name: {col_name}\n{df[col_name].unique()}\n')
  print(f'Index column\n{df.index.unique()}')
```

```python
print_unique_val_columns(case_df)
```

```python
case_df.replace('-', np.nan, inplace=True)

case_df.isna().any()
```

```python
case_df.index.isna().any()
```

```python
case_df.duplicated().any()
```

```python
case_df.columns.duplicated().any()
```

```python
case_df[(case_df['latitude'].isna()) \
        | (case_df['longitude'].isna()) \
        | (case_df['city'].isna())]
```

2) Patient Data
- **PatientInfo**: Epidemiological data of COVID-19 patients in South Korea

### PatientInfo data cleaning

Epidemiological data of COVID-19 patients in South Korea
- patient_id: the ID of the patient
  - patient_id(10) = region_code(5) + patient_number(5)
  - You can check the region_code in 'Region.csv'
  - There are two types of the patient_number
1) local_num: The number given by the local government.
2) global_num: The number given by the KCDC
- sex: the sex of the patient
- age: the age of the patient
  - 0s: 0 ~ 9
  - 10s: 10 ~ 19
  - ...
  - 90s: 90 ~ 99
  - 100s: 100 ~ 109
- country: the country of the patient
- province: the province of the patient
- city: the city of the patient
- infection_case: the case of infection
- infected_by: the ID of who infected the patient
  - This column refers to the 'patient_id' column.
- contact_number: the number of contacts with people
- symptom_onset_date: the date of symptom onset
- confirmed_date: the date of being confirmed
- released_date: the date of being released
- deceased_date: the date of being deceased
- state: isolated / released / deceased
  - isolated: being isolated in the hospital
  - released: being released from the hospital
  - deceased: being deceased

```python
date_parser = lambda date: pd.to_datetime(date,
                                          format='%Y/%m/%d',
                                          errors='coerce')

patient_info_df = pd.read_csv('PatientInfo.csv',
                              index_col=0,
                              parse_dates=[9,10,11,12],
                              date_parser=date_parser)
patient_info_df.head()
```

```python
print_summary_statistics(patient_info_df)
```

```python
patient_info_df.shape
```

```python
print_unique_val_columns(patient_info_df)
```

```python
patient_info_df.replace(['-',' '], np.nan, inplace=True)

patient_info_df.isna().any()
```

```python
patient_info_df.index.isna().any()
```

```python
drop_mask_1 = patient_info_df['contact_number'].isin(['1000000796',
                                                      '6100000098',
                                                      '6100000099'])

drop_mask_2 = patient_info_df['infected_by'].isin(['12702'])

patient_info_df.drop(patient_info_df[drop_mask_1 | drop_mask_2].index,
                     inplace=True)
```

```python
def print_nan_count_columns(df: pd.DataFrame) -> None:
  for col in df.columns:
    n_nan_val = df[col].isna().sum()
    print(f'Column name: {col}, number of NaN: {n_nan_val}')
```

```python
print_nan_count_columns(patient_info_df)
```

```python
patient_info_df[(patient_info_df['sex'].isna()) \
        & (patient_info_df['age'].isna()) \
        & (patient_info_df['infected_by'].isna()) \
        & (patient_info_df['contact_number'].isna())]
```

## Exploratory data analysis

### The most infected provinces and cities

```python
prov_by_conf = case_df.groupby(by='province')['confirmed'].sum()
prov_by_conf = prov_by_conf.sort_values(ascending=False)[:10]

fig, ax = plt.subplots(figsize=(12,8))

ax = sns.barplot(x=prov_by_conf.values,
                 y=prov_by_conf.index,
                 color='olive')

ax.set_xlabel('Confirmed cases')
ax.set_ylabel('Province')
ax.set_title('Top 10 provinces with highest number of confirmed cases',
             fontsize='x-large')

plt.show()
```

```python
city_by_conf = case_df.groupby(by='city')['confirmed'].sum()
city_by_conf = city_by_conf.sort_values(ascending=False)[:10]

fig, ax = plt.subplots(figsize=(12,8))

ax = sns.barplot(x=city_by_conf.values,
                 y=city_by_conf.index,
                 color='olive')

ax.set_xlabel('Confirmed cases')
ax.set_ylabel('City')
ax.set_title('Top 10 cities with highest number of confirmed cases',
             fontsize='x-large')

plt.show()
```

We can see that the most infected area is Daegu province and Nam-gu district which is a part of Daegu.

### The most common infection cases

```python
infec_case_counts = case_df['infection_case'].value_counts()[:10]

fig, ax = plt.subplots(figsize=(12,8))

ax = sns.barplot(x=infec_case_counts.values,
                 y=infec_case_counts.index,
                 color='olive')

ax.set_xlabel('Count')
ax.set_ylabel('Infection case')
ax.set_title('Most common infection cases',
             fontsize='x-large')

plt.show()
```

From the barplot above we can see that the most common cases of infection are unknown cases, but the following two are overseas inflow and contact with patient. Also, some other cases which involves highly crowded places like a church, clubs, call center. ### Death prediction

Creating the dataframe for training.

```python
death_pr_df = patient_info_df[patient_info_df.columns[0:]]
death_pr_df = death_pr_df[['sex','age','confirmed_date','released_date',
                           'deceased_date','state']]

death_pr_df.head()
```

```python
death_pr_df.dropna(subset=['sex','age','confirmed_date'], inplace=True)
death_pr_df.dropna(thresh=4, inplace=True)
```

```python
death_pr_df['confirmed_period_r'] = death_pr_df['released_date'] - death_pr_df['confirmed_date']
death_pr_df['confirmed_period_r'].fillna(pd.Timedelta(0, "d"), inplace=True)

death_pr_df['confirmed_period_d'] = death_pr_df['deceased_date'] - death_pr_df['confirmed_date']
death_pr_df['confirmed_period_d'].fillna(pd.Timedelta(0, "d"), inplace=True)

death_pr_df['confirmed_period'] = death_pr_df['confirmed_period_r'] + death_pr_df['confirmed_period_d']

death_pr_df['confirmed_period'].replace(pd.Timedelta(0, "d"),
                                        pd.NaT,
                                        inplace=True)

death_pr_df['confirmed_period'] = death_pr_df['confirmed_period'].dt.days

death_pr_df['confirmed_period'] = death_pr_df['confirmed_period'].apply(lambda x: -x if x < 0 else x)

death_pr_df['is_deceased'] = death_pr_df['state'].apply(lambda x: True if x=='deceased' else False)
death_pr_df['sex'] = death_pr_df['sex'].apply(lambda x: True if x=='male' else False)
death_pr_df['age'] = death_pr_df['age'].apply(lambda x: int(x[:-1]) if isinstance(x, str) else x)

death_pr_df.reset_index(inplace=True)

death_pr_df = death_pr_df[['sex','age','confirmed_period','is_deceased']]

death_pr_df.dropna(inplace=True)

death_pr_df.head()
```

```python
len(death_pr_df[~death_pr_df['is_deceased']])
```

Because there is only 60 cases where people die and 1505 where they don't, only 60 cases will be sampled from survived cases due to huge imbalance.

```python
x_y_df = death_pr_df[~death_pr_df['is_deceased']].sample(n=60)
x_y_df = pd.concat([x_y_df, death_pr_df[death_pr_df['is_deceased']]])
x_y_df.reset_index(drop=True, inplace=True)
```

```python
x_y_df[x_y_df['is_deceased']]
```

Normalize for regression model.

```python
x_df = x_y_df.iloc[:,[0,1,2]]
y_df = x_y_df.iloc[:,3]

scaler = StandardScaler()
scaler.fit(x_df)
x_df_scaled = scaler.transform(x_df)
x_df_scaled[:5]
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
```

Split into train and test datasets.

```python
x_train, x_test, y_train, y_test = train_test_split(x_df_scaled,
                                                    y_df,
                                                    test_size=0.2)
```

Train and test accuracy.

```python
model = LogisticRegression()
model.fit(x_train, y_train)
model.score(x_test, y_test)
```

Test on custom data.

```python
df = pd.DataFrame({'sex':[False],
                   'age':[30],
                   'confirmed_period':[20]})

x_scaled = scaler.transform(df)
prediction = model.predict(x_scaled)

if prediction[0]:
  print('A person is likely to die')
else:
  print('A person is likely to live')
```
